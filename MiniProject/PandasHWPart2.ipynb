```python
# Dependencies
import pandas as pd
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"


# Import the books.csv file as a DataFrame
bootcamp_df = pd.read_csv(csv_path, encoding ='iso-8859-1', low_memory=False)
bootcamp_df
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

reduced_df = bootcamp_df.iloc [:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_df
```

```python
reduced_df.dtypes
```

```python
# Change reading 0 and 1 to No and Yes, respectively
reduced_recode_df = reduced_df.replace({0.0: "No", 1.0: "Yes"})
reduced_recode_df

reduced_recode_df
```

```python
# Extract rows for only those who attended a bootcamp
attendees = reduced_recode_df.loc[reduced_recode_df["AttendedBootcamp"] == "Yes"]
attendees.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bc_name_count = attendees['BootcampName'].value_counts()
bc_name_count
```

```python

bc_df = pd.DataFrame ({"BootcampName" : ["General Assembly" , "Flatiron School", "Dev Bootcamp", "The Iron Yard", "Prime Digital Academy"]}, columns=["BootcampName"])
bc_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
#select columns for bootcamp name BootcampName, attendeed AttendedBootcamp, and recommend BootcampRecommend - all records

attendees_reduced = attendees[["BootcampName", "AttendedBootcamp", "BootcampRecommend"]]
attendees_reduced
```

```python
bc_group_df = attendees['BootcampName'].value_counts()
bc_group_df
```

```python
bc_group_recommend_df = attendees.groupby("BootcampName")
bc_group_recommend_df[["BootcampRecommend"]].count()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
left_merge_df = pd.merge(bc_df, attendees_reduced, on="BootcampName", how="left")
left_merge_df
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
bc_name_count_2 = left_merge_df['BootcampName'].value_counts()
bc_name_count_2
```
